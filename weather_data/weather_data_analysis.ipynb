```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Project+1+-+Weather+Dataset.csv')
df.head()
```

```python
df.dtypes
```

```python
# Convert Date/Time to datetime format
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
df.set_index('Date/Time', inplace=True)
df.dtypes
```

```python
df.info()
```

```python
df.describe()
```

```python
# Checking temperature trends over the time frame

plt.figure(figsize=(12,5))
plt.plot(df.index, df['Temp_C'], color='tab:red', alpha=0.7, label='Temperature (C)')
# plt.plot(df.index, df['Temp_C'], color='tab:red', alpha=0.7, label="Temperature (C)")
plt.xlabel('Time')
plt.ylabel('Temperature (C)')
plt.title('Temperature Trends over Time')
plt.legend()
plt.show()
```

```python
# Checking wind speed over the time frame

plt.figure(figsize=(12,13))
plt.plot(df.index, df['Wind Speed_km/h'], color='tab:blue', alpha=0.7, label='Wind Speed (Km/h)')
plt.xlabel('Time')
plt.ylabel('Wind Speed (Km/h)')
plt.title("Wind Speed Pattern over Time")
plt.legend()
plt.show()
```

```python
# Checking visibility over the time frame

plt.figure(figsize=(12,13))
plt.plot(df.index, df['Visibility_km'], color='tab:green', alpha=0.7, label='Visibility')
plt.xlabel('Time')
plt.ylabel('Visibility')
plt.title("Visibility Pattern over Time")
plt.legend()
plt.show()
```

```python
# Checking Humidity over the time frame

plt.figure(figsize=(12,13))
plt.plot(df.index, df['Rel Hum_%'], color='tab:green', alpha=0.7, label='Humidity (%)')
plt.xlabel('Time')
plt.ylabel('Humidity (%)')
plt.title("Humidity Pattern over Time")
plt.legend()
plt.show()
```

```python
# Visibility vs Humidity patterns
plt.figure(figsize=(12,5))
plt.plot(df.index, df['Visibility_km'], color='tab:green', alpha=0.7, label='Visibility')
plt.plot(df.index, df['Rel Hum_%'], color='tab:red', alpha=0.7, label='Humidity (%)')

plt.xlabel('Time')
plt.ylabel('Value')
plt.title("Visibility Vs Humidity Pattern over Time")
plt.legend()
plt.show()
```

```python
# Visibility vs Temperature patterns
plt.figure(figsize=(12,5))
plt.plot(df.index, df['Visibility_km'], color='tab:green', alpha=0.7, label='Visibility')
plt.plot(df.index, df['Temp_C'], color='tab:red', alpha=0.7, label='Temperature (C)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title("Visibility Vs Temperature Pattern over Time")
plt.legend()
plt.show()
```

```python
# Visibility vs Wind speed patterns
plt.figure(figsize=(12,5))
plt.plot(df.index, df['Visibility_km'], color='tab:green', alpha=0.7, label='Visibility')
plt.plot(df.index, df['Wind Speed_km/h'], color='tab:purple', alpha=0.7, label='Wind Speed (Km/h)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title("Visibility Vs Wind Speed Pattern over Time")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Between Weather Variables")
plt.show()
```
