ML Models

```python
# Evaluation metrics function
def evaluate_model(y_true, y_pred):
    from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    print(f'Mean Absolute Error: {mae}')
    print(f'Mean Squared Error: {mse}')
    print(f'Root Mean Squared Error: {rmse}')
    print(f'R^2 Score: {r2}')
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from lightgbm import LGBMRegressor
```

```python
x = df[
    [
        'Product',
        'City',
        'Manager',
        'Purchase Type',
        'Payment Method',
    ]
]

y = df['Revenue Generated']

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
# One hot encoding categorical variables
categorical_features = ['Product', 'City', 'Purchase Type', 'Payment Method', 'Manager']
# numerical_features = ['Price', 'Quantity']
```

```python
# The Preprocessor
preprocessor = ColumnTransformer(transformers=[
        # ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])
preprocessor
```

```python
#  Lightgbm pipeline
LGBMmodel = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LGBMRegressor())
])
LGBMmodel
```

```python
LGBMmodel.fit(x_train, y_train)
```

```python
lightgbm_predictions = LGBMmodel.predict(x_test)
print(f'-----------LightGBM Model Evaluation-----------')
evaluate_model(y_test, lightgbm_predictions)
```

```python
# lightgbm_predictions vs y_test plot
plt.figure(figsize=(10,6))
plt.scatter(y_test, lightgbm_predictions, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual Revenue Generated')
plt.ylabel('Predicted Revenue Generated')
plt.title('LightGBM: Actual vs Predicted Revenue Generated')
plt.show()
```

```python
# Random Forest pipeline
rf_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestRegressor(n_estimators=500,max_depth=20,min_samples_split=5,min_samples_leaf=2,max_features='sqrt',random_state=42,n_jobs=-1))
])
rf_pipeline
```

```python
rf_pipeline.fit(x_train, y_train)
```

```python
rf_predictions = rf_pipeline.predict(x_test)
print(f'-----------Random Forest Model Evaluation-----------')
evaluate_model(y_test, rf_predictions)
```

```python
# rf_predictions vs y_test plot
plt.figure(figsize=(10,6))
plt.scatter(y_test, rf_predictions, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual Revenue Generated')
plt.ylabel('Predicted Revenue Generated')
plt.title('Random Forest: Actual vs Predicted Revenue Generated')
plt.show()
```

```python
# Linear regression pipeline
linreg_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LinearRegression())
])
linreg_pipeline
```

```python
linreg_pipeline.fit(x_train, y_train)
```

```python
linregPredictions = linreg_pipeline.predict(x_test)
print(f'-----------Linear Regression Model Evaluation-----------')
evaluate_model(y_test, linregPredictions)
```

```python
# linear regression predictions vs y_test plot
plt.figure(figsize=(10,6))
plt.scatter(y_test, linregPredictions, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual Revenue Generated')
plt.ylabel('Predicted Revenue Generated')
plt.title('Linear Regression: Actual vs Predicted Revenue Generated')
plt.show()
```
