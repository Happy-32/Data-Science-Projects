```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Housing+Data.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['houses_sold'].fillna(df['houses_sold'].median(), inplace=True)
```

```python
df['no_of_crimes'].fillna(0, inplace=True)
# 0 indicates no crimes reported
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['date']=pd.to_datetime(df['date'])
df.set_index(df['date'], inplace=True)
df.dtypes
```

```python
plt.figure(figsize=(12,5))
plt.plot(df.index, df['houses_sold'], color='tab:red', label='Houses Sold')
plt.xlabel('Time')
plt.ylabel('Houses Sold')
plt.title('Houses Sold over Time')
plt.legend()
plt.show()
```
